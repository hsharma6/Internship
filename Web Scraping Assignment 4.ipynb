# Q1: Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]

try:
    rnk=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for i in rnk[0:30]:
        rank.append(i.text.split('.')[0])
except NoSuchElementException:
    rank.append('-')
    
try:
    nm=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
    for i in nm[0:30]:
        name.append(i.text.split('[')[0])
except NoSuchElementException:
    name.append('-')
    
try:
    art=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
    for i in art[0:30]:
        artist.append(i.text.split('-')[0])
except NoSuchElementException:
    artist.append('-')
    
try:
    updt=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    for i in updt[0:30]:
        upload_date.append(i.text)
except NoSuchElementException:
    upload_date.append('-')
    
try:
    vws=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for i in vws[0:30]:
        views.append(i.text)
except NoSuchElementException:
    views.append('-')
```

```python
df1=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Upload Date':upload_date,'Views(Billions)':views})
df1
```

# Q8: Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
search=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
search.click()
```

```python
dataset_url=[]

url=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in url:
    dataset_url.append(i.get_attribute('href'))
```

```python
len(dataset_url)
```

```python
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
no_of_instances=[]
no_of_attribute=[]
year=[]

for i in dataset_url:
    driver.get(i)
    try:
        dn=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        dataset_name.append(dn.text)
    except NoSuchElementException:
        dataset_name.append('-')
        
    try:
        dt=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[1]/p')
        data_type.append(dt.text)
    except NoSuchElementException:
        data_type.append('-')
        
    try:
        tsk=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        task.append(tsk.text)
    except NoSuchElementException:
        task.append('-')
        
    try:
        at=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        attribute_type.append(at.text)
    except NoSuchElementException:
        attribute_type.append('-')
        
    try:
        noi=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        no_of_instances.append(noi.text)
    except NoSuchElementException:
        no_of_instances.append('-')
        
    try:
        noa=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        no_of_attribute.append(noa.text)
    except NoSuchElementException:
        no_of_attribute.append('-')
        
    try:
        yr=driver.find_element(By.XPATH,'//h2[@class="text-primary-content"]')
        year.append(yr.text)
    except NoSuchElementException:
        year.append('-')
```

```python
df8=pd.DataFrame({'Dataset Name':dataset_name,'Data Type':data_type,'Task':task,'Attribute Type':attribute_type,'No of instances':no_of_instances,'No of attribute':no_of_attribute,'Year':year})
df8
```

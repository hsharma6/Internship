# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
driver.get("https://www.naukri.com/")
```

```python
title=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title[0:10]:
    result=i.text
    job_title.append(result)
```

```python
location=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in location[0:10]:
    result1=i.text
    job_location.append(result1)
    
company=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company[0:10]:
    result2=i.text
    company_name.append(result2)
    
experience=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in experience[0:10]:
    result3=i.text
    experience_required.append(result3)
```

# Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes: Brand, Product Description, Price.

```python
product=driver.find_element(By.CLASS_NAME,'Pke_EE')
product.send_keys('Sunglasses')
```

```python
search=driver.find_element(By.CLASS_NAME,'_2iLD__')
search.click()
```

```python
brand=[]
product_description=[]
price=[]
```

```python
start=0
end=3
for page in range(start,end):
    brands=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brands:
        brand.append(i.text)
        
    prdt_des=driver.find_elements(By.CLASS_NAME,'IRpwTa')
    for i in prdt_des:
        product_description.append(i.text)
    
    prices=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in prices:
        price.append(i.text)
        
next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
next_button.click()

brand_result=brand[0:100]
product_description_result=product_description[0:100]
price_result=price[0:100]
```

```python
df3=pd.DataFrame({'Brand':brand_result,'Product Description':product_description_result,'Price':price_result})
df3
```

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time. The above task will be done in following steps: 1. First get the webpagehttps://www.azquotes.com/ 2. Click on TopQuotes 3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
driver.get("https://www.azquotes.com/")
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
search.click()
```

```python
quotes=[]
authors=[]
type_of_quotes=[]
```

```python
start=0
end=10

for page in range(start,end):
    quote=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote:
        quotes.append(i.text)
        
    author=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author:
        authors.append(i.text)
        
    type_of_quote=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_of_quote:
        type_of_quotes.append(i.text)    
    
next_button=driver.find_element(By.XPATH,'//li[@class="next"]')
next_button.click()
```

```python
df8=pd.DataFrame({'Quotes':quotes,'Authors':authors,'Type of Quotes':type_of_quotes})
df8
```

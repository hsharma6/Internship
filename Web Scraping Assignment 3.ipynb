and save it in a data frame and csv. In case if any product has less than 3 pages in search results then
scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 


```python
driver.get("https://www.amazon.in/")
```

```python
search_bar=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search_bar.send_keys('Guitar')

search_button=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search_button.click()
```

```python
guitar_url=[]
start=0
end=3

for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        guitar_url.append(i.get_attribute('href'))
next_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
next_button.click()
```

```python
len(guitar_url)
```

```python
brand=[]
product_name=[]
price=[]
replacement=[]
expected_delivery=[]
availability=[]
for i in guitar_url:
    driver.get(i)
    try:
        brnd=driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        brand.append(brnd.text)
    except NoSuchElementException:
        brand.append('-')   
        
    try:
        prdct_nm=driver.find_element(By.XPATH,'//span[@id="productTitle"]')
        product_name.append(prdct_nm.text)
    except NoSuchElementException:
        product_name.append('-')
        
    try:
        prc=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        price.append(prc.text)
    except NoSuchElementException:
        price.append('-')
        
    try:
        replace=driver.find_element(By.XPATH,'//*[@id="iconfarmv2_feature_div"]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/ol/li[3]/div[1]/span/div[2]/span')
        replacement.append(replace.text)
    except NoSuchElementException:
        replacement.append('-')
        
    try:
        exp_del=driver.find_element(By.XPATH,'//*[@id="deliveryBlockContainer"]/div[1]/div[1]/div[1]/div[1]/span/span[1]')
        expected_delivery.append(exp_del.text)
    except NoSuchElementException:
        expected_delivery.append('-')
        
    try:
        avail=driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-none a-spacing-top-micro }"]')
        availability.append(avail.text)
    except NoSuchElementException:
        availability.append('-')
```

```python
df=pd.DataFrame({'Brand Name':brand,'Product Name':product_name,'Price':price,'Return/Exchange':replacement,'Expected Delivery':expected_delivery,'Availability':availability,'Product URL':guitar_url})
df
```

```python
brand_name=[]
smartphone_name=[]
colour=[]
ram=[]
rom=[]
primary_camera=[]
secondary_camera=[]
display_size=[]
battery_capacity=[]
price=[]
for i in phone_url:
    driver.get(i)
    try:
        brnd_nm=driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        brand_name.append(brnd_nm.text.split(' ')[0])
    except NoSuchElementException:
        brand_name.append('-')   
    
    try:
        smrtphn_nm=driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        smartphone_name.append(smrtphn_nm.text.split('(')[0])
    except NoSuchElementException:
        smartphone_name.append('-') 
        
    try:
        clr=driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        colour.append(clr.text.split('(')[1].split(',')[0])
    except NoSuchElementException:
        colour.append('-')  
        
    try:
        rm=driver.find_element(By.XPATH,'//*[@class="_2418kt"]/ul/li')
        ram.append(rm.text.split('|')[0])
    except NoSuchElementException:
        ram.append('-') 
        
    try:
        rm1=driver.find_element(By.XPATH,'//*[@class="_2418kt"]/ul/li')
        rom.append(rm1.text.split('|')[1])
    except NoSuchElementException:
        rom.append('-')  
        
    try:
        prm_cm=driver.find_element(By.XPATH,'//*[@class="_2418kt"]/ul/li[3]')
        primary_camera.append(prm_cm.text.split('|')[0])
    except NoSuchElementException:
        primary_camera.append('-')  
        
    try:
        sec_cm=driver.find_element(By.XPATH,'//*[@class="_2418kt"]/ul/li[3]')
        secondary_camera.append(sec_cm.text)
    except NoSuchElementException:
        secondary_camera.append('-')
        
    try:
        dis_sz=driver.find_element(By.XPATH,'//*[@class="_2418kt"]/ul/li[2]')
        display_size.append(dis_sz.text.split('(')[0])
    except NoSuchElementException:
        display_size.append('-') 
        
    try:
        dis_sz=driver.find_element(By.XPATH,'//*[@class="_2418kt"]/ul/li[4]')
        battery_capacity.append(dis_sz.text)
    except NoSuchElementException:
        battery_capacity.append('-')
        
    try:
        prc=driver.find_element(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')
        price.append(prc.text)
    except NoSuchElementException:
        price.append('-')  
```

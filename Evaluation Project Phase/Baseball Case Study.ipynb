# Implementing Machine Learning thru Linear Regression

```python
x=df.iloc[:,1:16]
y=df.iloc[:,0]
```

Seperated target and its features so that x only has input variables and y only has output variable

```python
x
```

```python
y
```

Creating train test split

Training phase will get 70% of distribution and testing phase will get 30% of distribution.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
LR.score(x_train,y_train)
```

The above score shows that the model performed really good in its traing phase.

```python
LRpred=LR.predict(x_test)
print("Predicted value", LRpred)
print("Actual value", y_test)
```

The above data shows the predicted value and the actual value.

```python
print('Mean Squared Error:', MSE(y_test, LRpred))
print('Mean Absolute Error:', MAE(y_test, LRpred))

print('Root Mean Squared Error:', np.sqrt(MSE(y_test, LRpred)))

print('R2 Score:', r2_score(y_test, LRpred))
```

The model has an Root Mean Squared Error score of ~4.86. This implies that lower is the RMSE score, therefore model is performing better at making predictions.

# Using Random Forest Regressor

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)
```

```python
RFR.score(x_train,y_train)
```

The above score shows that the model did good in its traing phase.

```python
RFRpred=RFR.predict(x_test)
print("Predicted value", RFRpred)
print("Actual value", y_test)
```

The above data shows the predicted value and the actual value. We can see that there is not much difference b/w both the data so the model is working good in predicting Wins.

```python
print('Mean Squared Error:', MSE(y_test, RFRpred))
print('Mean Absolute Error:', MAE(y_test, RFRpred))

print('Root Mean Squared Error:', np.sqrt(MSE(y_test, RFRpred)))

print('R2 Score:', r2_score(y_test, RFRpred))
```

The model has an Root Mean Squared Error score of ~7.29. This implies that lower is the RMSE score, therefore model is performing better at making predictions.
